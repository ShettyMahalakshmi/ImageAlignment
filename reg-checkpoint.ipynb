```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import scipy
```

```python
MAX_FEATURES = 500

GOOD_MATCH_PERCENT = 0.2

img_ref = cv2.imread("C:/AM/Imageprocessing/ref.jpg",1)
rows,cols,channels = tuple(img_ref.shape)
dim = (cols, rows)
# resize image

img_test = cv2.imread("C:/AM/Imageprocessing/4.jpg",1)
img_test = cv2.resize(img_test, dim, interpolation = cv2.INTER_AREA)

#print(rows)
#print(cols)
print(img_ref.shape)
print(img_test.shape)
```

```python
 #Convert images to grayscale
img_testGray = cv2.cvtColor(img_test, cv2.COLOR_BGR2GRAY)
img_refGray = cv2.cvtColor(img_ref, cv2.COLOR_BGR2GRAY)
   
# Detect ORB features and compute descriptors.
orb = cv2.ORB_create(MAX_FEATURES)
keypoints1, descriptors1 = orb.detectAndCompute(img_testGray, None)
keypoints2, descriptors2 = orb.detectAndCompute(img_refGray, None)

# Match features.
matcher = cv2.DescriptorMatcher_create(cv2.DESCRIPTOR_MATCHER_BRUTEFORCE_HAMMING)
matches = matcher.match(descriptors1, descriptors2, None)

# Sort matches by score

matches.sort(key=lambda x: x.distance, reverse=False)

# Remove not so good matches
numGoodMatches = int(len(matches) * GOOD_MATCH_PERCENT)
matches = matches[:numGoodMatches]

# Extract location of good matches
points1 = np.zeros((len(matches), 2), dtype=np.float32)
points2 = np.zeros((len(matches), 2), dtype=np.float32)
for i, match in enumerate(matches):
    points1[i, :] = keypoints1[match.queryIdx].pt
    points2[i, :] = keypoints2[match.trainIdx].pt
# Find homography
h, mask = cv2.findHomography(points1, points2, cv2.RANSAC)
# Use homography
height, width, channels = img_test.shape
im1Reg = cv2.warpPerspective(img_test, h, (width, height))
```
